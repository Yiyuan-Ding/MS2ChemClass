```python
import pickle
from matplotlib import pyplot as plt
```

## NPClassifier LDA

```python
with open('/mnt/scratch/ding013/MS2ChemClass/accuracy_per_label2/accuracy_per_label_npc_class_lda_min4.pickle', 'rb') as f:
    class_accuracy = pickle.load(f)
with open('/mnt/scratch/ding013/MS2ChemClass/accuracy_per_label2/accuracy_per_label_npc_class_lda_without_multilabels_min4.pickle', 'rb') as f:
    class_accuracy_without_multilabel = pickle.load(f)
data_type = 'npc_class'    
save_path = f'/mnt/scratch/ding013/MS2ChemClass/scatter_accuracy_vs_groupsize/scatter_accuracy_with_and_without_multilables_{data_type}.svg'
a1= class_accuracy['accuracy']
a2= class_accuracy_without_multilabel['accuracy']
lbs1 = class_accuracy['labels']
lbs2 = class_accuracy_without_multilabel['labels']
new_a1=[]
new_a2=[]
print(len(a1),len(a2),len(lbs1),len(lbs2))
for i,v in enumerate(lbs2):
    new_a1.append(a1[lbs1.index(v)])
    new_a2.append(a2[i])

plt.rcParams.update({'figure.figsize':(10,7), 'figure.dpi':100})
plt.scatter(new_a1, new_a2)
plt.plot((0, 1), "r--")
plt.title(f'accuracy with multiple-label groups  vs. accuracy without multiple-label groups in {data_type}')
plt.xlabel('accuracy with multiple-label groups')
plt.ylabel('accuracy without multiple-label groups')
for i,txt in enumerate(lbs2):
    if abs(new_a2[i]-new_a1[i])>0.5:
        #plt.annotate(txt,(new_a1[i],new_a2[i]))
        print(txt,(new_a1[i],new_a2[i]))
plt.savefig(save_path)
plt.show()
```
